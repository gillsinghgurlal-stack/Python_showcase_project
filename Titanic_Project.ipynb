# INTRODUCTION ABOUT DATASET

The Titanic dataset available in the Seaborn library (sns.load_dataset("titanic")) is a popular dataset used for data analysis. It contains information about passengers aboard the Titanic.
This dataset includes details such as passenger class (pclass), gender (sex), age (age), fare paid (fare), embarkation port (embarked), and survival status (survived).
The primary objective when working with this dataset is to analyze the factors that influenced passenger survival. 
Because of its mix of categorical and numerical variables, the Titanic dataset is ideal for beginners to understand real-world data analysis and predictive modeling in Python.

# DATA CLEANING (Sumit Gill)

# Checking no of duplicated rows

```python
df.duplicated().sum()
```

# Deleting duplicate rows

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

# Droping Unnecessary column and column having large null values and low significance 

```python
df=df.drop(columns=['sibsp','parch','embarked','who','adult_male','alone','deck'])
```

```python
df.info()
```

# Filling null values with unknown

```python
df['embark_town'] = df['embark_town'].fillna('unknown')
```

```python
df.info()
```

# Filling Null Values in Age Column with mean value of column to maintain the Data integrity

```python
df['age'] = df['age'].fillna(df['age'].mean())
```

# Comparison of survived and non survived

```python
(df['survived'].value_counts()).round(2)
```

```python
(df['survived'].value_counts(normalize=True) * 100).round(2)
```

```python
((df['survived'].value_counts(normalize=True) * 100).round(2)).plot(kind='bar')
plt.ylabel("percentage")
```

# OBSERVATIONS BY GROUPING (Prince Shakya)

# OBSERVATION:How many male and female are survived in that incindent from the total passenger followed by Gender column.

```python
survival_by_sex = df.groupby('sex')['survived'].agg(
    Total_Passengers='count',
    Total_Survived='sum',
    Survival_mean='mean'
)
```

```python
survival_by_sex
```

# OBSERVATION:How many people are survived by class from the total passengers.

```python
survival_by_class = df.groupby('class')['survived'].agg(
    Total_Passengers='count',
    Total_Survived='sum',
    Survival_mean='mean'
)
```

```python
survival_by_class
```

# OBSERVATION:how many passengers are survived in differnt city from the total passengrs.

```python
survival_by_boarding_city = df.groupby('embark_town')['survived'].agg(
    Total_Passengers='count',
    Total_Survived='sum',
    Survival_mean='mean'
)
```

```python
survival_by_boarding_city
```

# Observation: How many people were alive after the incident.

```python
total_alive=df.groupby('alive')['sex'].agg(
    total_passengers='count'
)
```

```python
total_alive
```

OBSERVATION: We are creating the lables of different age group and checking how many people are are survived in the different age group.



```python
df['age_group'] = pd.cut(
    df['age'],
    bins=[0, 12, 18, 35, 60, 100],
    labels=['Child', 'Teen', 'Young Adult', 'Adult', 'Senior']
)

print(df[['age', 'age_group']].head(100))
```

```python
age_group_survive= df.groupby('age_group')['survived'].agg(
    Total_Passengers='count',
    Total_Survived='sum',
    survival_mean='mean'
)
```

```python
age_group_survive
```

# Visualization (Shubham Sain)

```python
df.groupby('sex')['survived']
```

```python
sns.set(style='whitegrid')
```

```python
plt.figure()
sns.barplot(x='sex', y='survived', data=df)

plt.title('Survival Rate by Gender')
plt.ylabel('Survival Rate')
plt.xlabel('Gender')
plt.show()
```

```python
#Female passengers had a significantly higher survival rate than male passengers.
#this suggests that rescue priority might have been given based on gender during evacuation.
```

```python
plt.figure()
sns.barplot(x='class', y='survived', data=df)

plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
# First-class passengers had the highest survival rate.
# Third-class passengers had the lowest survival probability.
#Socio-economic status may have influenced access to lifeboats.
```

```python
plt.figure()
sns.countplot(x='embark_town', hue='survived', data=df)

plt.title('Survival Count by Embark Town')
plt.xlabel('Boarding City')
plt.ylabel('Passenger Count')
plt.show()
```

```python
#Passengers who boarded from Cherbourg showed better survival rates.
#This may relate to higher-class ticket holders embarking from that port.
```

```python
plt.figure()
sns.barplot(x='age_group', y='survived', data=df)

plt.title('Survival Rate by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Survival Rate')
plt.show()
```

```python
#Children had relatively higher survival rates compared to adults.
# This suggests priority may have been given to younger passengers.
```

```python
sns.scatterplot(data=df, x='age', y='fare',hue='pclass',palette='Set1')

plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```
